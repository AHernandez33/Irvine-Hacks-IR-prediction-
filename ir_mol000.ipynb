```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV, ensure numeric data
df = pd.read_csv('200mol.csv', header=0)        # adjust header row if needed
df = df.apply(pd.to_numeric, errors='coerce')   # convert all to numbers

# Extract first molecule (skip name column if present)
first_molecule = df.iloc[71, 1:].values  # use 0,: if no name column

# Generate wavenumbers (4000  400 cm)
wavenumbers = np.linspace(4000, 400, len(first_molecule))

# Plot IR spectrum
plt.figure(figsize=(10,6))
plt.plot(wavenumbers, first_molecule, color='blue', linewidth=1.5)
plt.gca().invert_xaxis()
plt.title('IR Spectrum of First Molecule', fontsize=16)
plt.xlabel('Wavenumber (cm)', fontsize=14)
plt.ylabel('Transmittance', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
